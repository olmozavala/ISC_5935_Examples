```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 3
x = np.linspace(0,10,100)
y = x*3 + 3 - np.random.rand(len(x))*6
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))
ax .plot(x,y, '.')
plt.show()
```

```python
def cost(mod_y,y):
    return np.mean(np.sum((mod_y-y)**2))
```

```python
m_options = np.linspace(0,6,100)
cost_curve = [cost(x*m, y) for m in m_options]
fig, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].plot(x, y, '.')
axs[1].plot(m_options, cost_curve, 'r')
# ------- Computing for a batch only --------
# Index to use
idx = np.arange(len(x))
for i in range(10):
    np.random.shuffle(idx)
    batch_size = 5
    new_x = x[idx][:batch_size]
    new_y = y[idx][:batch_size]
    new_x.sort()
    new_y.sort()
    axs[2].plot(m_options, [cost(new_x*m, new_y) for m in m_options], '--')

axs[0].grid()
axs[0].set_title("Data")
axs[1].grid()
axs[1].set_title("Cost function of all data")
axs[2].grid()
axs[2].set_title("Cost function of random batches of size 10")
plt.suptitle("Optimal curves for diffent batches")
plt.show()
```
